```python
fetcher = Fetcher('34.250.225.205', dir_raw_cache='data/test_tmp')
# fetcher.update_deals()
# fetcher.update_md_batch(fetch_new=False, parse_new=True, max_batch_size_gb=0.75, num_processes=8, book_type='tob', depth=20)

df = fetcher.fetch_qdb('feed_kraken_tob_5', '2025-10-26T00:00:00.000000Z', '2025-10-26T00:14:59.999999Z')
print(df.head())

_name = df.name.unique()[42]
df_plot = df[df.name == _name]
df_plot_indexed = df_plot.set_index('ts_server')

# Create curves for ask and bid
ask_curve = hv.Curve(df_plot_indexed, 'ts_server', 'ask_px_0', label='Ask').opts(color='red')
bid_curve = hv.Curve(df_plot_indexed, 'ts_server', 'bid_px_0', label='Bid').opts(color='green')


# Overlay them
plot = hv.Overlay([ask_curve, bid_curve]).opts(
    width=800,
    height=400,
    xlabel='Time',
    ylabel='Price',
    title=f'[Bid/Ask] {_name}',
    legend_position='top_right'
)
hv.save(plot, 'curve.html')
plot
```
